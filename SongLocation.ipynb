```python
def add_marker_with_link(map_obj, location, popup_text, link_url):
    popup_html = f'<p>{popup_text}</p><p><a href="{link_url}" target="_blank">link to the music</a></p>'
    popup = folium.Popup(popup_html, max_width=300)
    
    folium.Marker(location=location, popup=popup).add_to(map_obj)



my_map = folium.Map(location=[51.5244478,-0.1046743], zoom_start=10)


for item in modified_data:
    add_marker_with_link(my_map, item["location"], item["popup_text"], item["link_url"])

# Save the map to an HTML file
#my_map.save("osm_map_with_links.html")
my_map


```
